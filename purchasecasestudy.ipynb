# 🧩 Case Study: Understanding Consumer Purchasing Behavior in an E-Commerce Festival Sale
📌 **Problem Statement**

An e-commerce company has conducted a large-scale festival sale and collected detailed transaction-level data from over 550,000 purchase records. The business wants to understand what drives customer purchases, identify key customer segments, and uncover which product categories perform best across different demographics.

Your task in this case study is to analyze this dataset to answer the following core business question:

“What customer and product characteristics most strongly influence purchase amounts during the festival sale?”

By exploring patterns across gender, age groups, occupations, city categories, and product preferences, the goal is to help the company improve future marketing strategies, personalize recommendations, and optimize inventory planning.

📊** Dataset Explanation**

The dataset contains 550,068 rows and 12 columns, where each row represents a single transaction made by a user during the sale. Below is a clear explanation of each feature:

**Column	Description**

**User_ID**	Unique ID assigned to each customer (integer). Helps track multiple purchases by the same user.

**Product_ID**	Unique ID for each product purchased.

**Gender**	Customer gender (‘M’ or ‘F’). Useful for gender-based segmentation.

**Age**	Age group of the customer (e.g., ‘18–25’, ‘26–35’, etc.).

**Occupation**	Encoded integer representing customer’s occupation category.

**City_Category**	City tier where the customer resides (‘A’, ‘B’, or ‘C’).

Stay_In_Current_City_Years	How long the customer has lived in the current city (e.g., ‘1’, ‘2’, ‘4+’).

**Marital_Status**	Customer marital status (0 = Single, 1 = Married).

**Product_Category_1**	Main product category purchased (integer-coded).

**Product_Category_2**	Secondary product category (nullable — many products don’t have one).

**Product_Category_3**	Tertiary product category (also nullable).

**Purchase**	Purchase amount spent by the customer in that transaction. This is the key target variable for analysis.

#**Statistics**
Q1) it was observed that the average purchase made by the men of the age 18-25 was 10000.Is it still same?

```python
df1['Age'].value_counts()
```

```python
data=df1[(df1['Age']==1) & (df1['Gender']==1)]
```

```python
data.head()
```

```python
data['Purchase'].mean()
```

```python
sample=data.sample(3600,random_state=1)
sample.head()
```

#null hypothesis=mean will be equal to 10k
#alternative=mean will not be equal to 10k

```python
#one sample t-test
from scipy.stats import ttest_1samp
```

```python
pop_mean=10000
```

```python
t_stat,p_value=ttest_1samp(sample['Purchase'],pop_mean)
print(t_stat,p_value)
if p_value<0.05:
  print('we reject the null hypothesis')
else:
  print('we accept the null hypothesis')
```

#**Q2 is the average purchase made by men and women of the age 18-25 same?**

```python
data_men_sample=df1[(df1['Gender']==1) & (df1['Age']==1)]
data_women_sample=df1[(df1['Gender']==0) & (df1['Age']==1)]
```

```python
data_men_sample.shape
```

```python
data_women_sample.shape
```

```python
data_men_sample['Purchase'].mean()
```

```python
data_women_sample['Purchase'].mean()
```

```python
from scipy.stats import f
```

```python
F=data_men_sample.Purchase.mean()/data_women_sample.Purchase.mean()
F
```

```python
dff1=len(data_men_sample)-1
dff2=len(data_women_sample)-1
print(dff1,dff2)
```

```python
import scipy
p_val=1-scipy.stats.f.cdf(F, dff1, dff2)
print(p_val)
if p_val<0.05:
  print('we reject the null hypothesis')
else:
  print('we accept the null hypothesis')
```
